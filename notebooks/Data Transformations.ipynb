```python
from model.features import Features
```

```python
logFeatures = Features(transform='log')
normFeatures = Features(transform='norm')
rawFeatures = Features(transform=None)
qantileFeatures = Features(transform='quantile')
```

```python
metrics = ['Ozone', 'SO2', 'CO', 'NO2', 'PM25', 'PM10', 'Wind','Pressure', 'Temperature', 'Humidity', 'Fire']
```

```python
for tran, features in zip(['Log', 'Normed', 'Raw', 'Quantile'], [logFeatures, normFeatures, rawFeatures, qantileFeatures]):

    fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(12, 6))

    for metric, ax in zip(metrics, axes.flatten()):

        data = features.get_data(metric).values    
        ax.plot(data[:, 0], lw=0.5)

        ax.set_title(metric)
        ax.set_xticks([])

    axes[-1, -1].set_xticks([])
    axes[-1, -1].set_yticks([])
    
    plt.suptitle(tran)
    plt.tight_layout()
```

```python
for tran, features in zip(['Log', 'Normed', 'Raw', 'Quantile'], [logFeatures, normFeatures, rawFeatures, qantileFeatures]):

    fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(12, 6))

    for metric, ax in zip(metrics, axes.flatten()):

        data = features.get_data(metric).values    
        
        ax.hist(data.flatten(), bins=100)

        ax.set_title(metric)
        ax.set_xticks([])
        ax.set_yticks([])

    axes[-1, -1].set_xticks([])
    axes[-1, -1].set_yticks([])
    
    plt.suptitle(tran)
    plt.tight_layout()
```
